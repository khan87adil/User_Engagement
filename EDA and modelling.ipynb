```python
def dummy(df,col):
    drop = np.unique(df[col])[-1]
    dummy = pd.get_dummies(df[col])
    new_col = []
    for i in dummy.columns:
        new_col.append(col+'_'+str(i))
    dummy.columns = new_col
    df = pd.concat([df, dummy], axis=1)
    df.drop([col, new_col[-1]], inplace=True, axis=1)
    return df
```

```python
empty = []
for col in cols:
    try:
        df = dummy(df,col)
    except:
        empty.append(col)
```

# thresold value=0.06

```python
df_sub = df.iloc[:,(df_scaled1.var() > 0.06).values]
```

```python
df_sub.info()
```

```python
print(df_sub.columns)
```

```python
#df_sub.to_csv('Final_Dataset.csv')
```

```python
df_sub.shape
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt


# k means determine k
distortions = []
K = range(1,20)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(df_sub)
    kmeanModel.fit(df_sub)
    distortions.append(sum(np.min(cdist(df_sub, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / df_sub.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()

```

```python
kmeanModel = KMeans(n_clusters=5).fit(df_sub)
kmeanModel.fit(df_sub)
```

```python
df_sub['cluster'] = kmeanModel.labels_
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_sub.drop(["cluster"],1),df_sub["cluster"],test_size=0.2,random_state=7)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(x_train, y_train)
```

```python
print("MN Accuracy: %0.2f%%" % (100 * clf.score(x_test, y_test)))
```

```python
y_preds = clf.predict(x_test)
y_preds
```

# 3rd try ANN

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
#model.add(Dense(16, input_dim=len(features), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50, batch_size=100)
```

```python
in_sample_preds = [round(x[0]) for x in model.predict(x_train)]
out_sample_preds = [round(x[0]) for x in model.predict(x_test)]
```

```python
print('In-Sample Accuracy: %0.4f' % accuracy_score(y_train, in_sample_preds))
print('Out-of-Sample Accuracy: %0.4f' % accuracy_score(y_test, out_sample_preds))

print('\n')

print('In-Sample Precision: %0.4f' % precision_score(y_train, in_sample_preds,average="weighted"))
print('Out-of-Sample Precision: %0.4f' % precision_score(y_test, out_sample_preds,average="weighted"))

print('\n')

print('In-Sample Recall: %0.4f' % recall_score(y_train, in_sample_preds,average="weighted"))
print('Out-of-Sample Recall: %0.4f' % recall_score(y_test, out_sample_preds,average="weighted"))
```
